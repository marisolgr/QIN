```python
# all species
import xarray as xr
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.simplefilter('ignore') # filter some warning messages
```

```python
fishes = ['Sablefish','Dungeness Crab','Hake','Lingcod','P. Halibut','Chinook','Coho','Steelhead']
var = ['thetao','o2']   
depth = ['surface','bottom']
ssps = ['ssp485','ssp585']
Trange = np.array([[4.72, 12.85],[5.36,11.72],[1.18,23.29],[6.56,13],[2.92,9.28],[3.49,11.65],[4.43,12.35],[6.38,16]])
O2range = np.array([[35.13, 298.72],[107.91,305.91],[111,169],[33.55,299.94],[55.67,319.55],[54.42,309.15],[58.94,316],[94.27,306.19]])
O2range = O2range/44.661  # conversion to ml/l
Deprange = np.array([[175, 1450],[41,90],[126,229],[18,60],[155,515],[43,172],[28,118],[22,95]])
```

```python
sidx = np.argsort(Trange[:,1])
orfish = list()
for i in sidx:
    orfish.append(fishes[i])

plt.figure(figsize=(6,4))
for idx, i in enumerate(sidx):
    plt.plot([Trange[i,0],Trange[i,1]],[idx,idx],'-', lw=20, c='tab:orange')
plt.grid(True,alpha=0.3)
plt.xlabel('Optimal Range Temperature (C)', fontsize=14)
plt.yticks(range(len(fishes)),orfish, fontsize=14)
plt.ylim([-0.6, len(fishes)-0.4])
plt.show()


plt.figure(figsize=(6,2))
i=0
plt.plot([Trange[i,0],Trange[i,1]],[0,0],'-', lw=25, c='tab:orange')
plt.grid(True,alpha=0.3)
plt.xlabel('Optimal Range Temperature (C)', fontsize=14)
plt.yticks([0],['Sablefish'], fontsize=14)
plt.ylim([-0.1, +0.1])
plt.xlim([3,14])
plt.show()
```

```python
sidx = np.argsort(O2range[:,0])
orfish = list()
for i in sidx:
    orfish.append(fishes[i])

plt.figure(figsize=(6,4))
for idx, i in enumerate(sidx):
    plt.plot([O2range[i,0],O2range[i,1]],[idx,idx],'-', lw=20, c='tab:blue')
plt.axvline(x=2,ls='--',c='k',lw=2)
plt.grid(True,alpha=0.3)
plt.xlabel('Optimal Range Oxygen (ml/L)', fontsize=14)
plt.yticks(range(len(fishes)),orfish, fontsize=14)
plt.ylim([-0.6, len(fishes)-0.4])
plt.show()
    
plt.figure(figsize=(6,2))
i=0
plt.plot([O2range[i,0],O2range[i,1]],[0,0],'-', lw=25, c='tab:blue')
plt.grid(True,alpha=0.3)
plt.xlabel('Optimal Range O2 (C)', fontsize=14)
plt.yticks([0],['Sablefish'], fontsize=14)
plt.ylim([-0.1, +0.1])
plt.xlim([0,10])
plt.show()
```
