# Species

```python
fishes = ['Sablefish','Anchovy']
Trange = np.array([[4.72, 12.85]])
O2range = np.array([[35.123, 400]])
O2range = O2range/44.661  # conversion to ml/l
Deprange = np.array([[175, 1450]] )
```

```python
def suitable_habitat_qin_decadal_map(fish):
    for i,ina in enumerate(fishes):
        if 'Sablefish' in ina:
            fishnum=i
    
    # decades to average
    yrsav = [*range(1990,2100,10)]
    var = ['thetao','o2']   
    depth = ['surface','bottom']
    
    for i2 in depth:
        # for each variable at given depth
        for i in var:
            tmp1 = np.full((len(qin.lat.values),len(qin.lon.values),len(yrsav),9),np.nan)
            # optimal range thresholds
            if i=='thetao':
                th1 = Trange[fishnum,0]
                th2 = Trange[fishnum,1]
            else:
                th1 = O2range[fishnum,0]
                th2 = O2range[fishnum,1]
            # for each model
            fns = glob.glob(ldp+'/QIN/downscaled_data/'+i2+i+'/*.nc')
            # build datasets from all models
            for modn,fn in enumerate(fns):
                # model number
                for k,kna in enumerate(models):
                    if kna in fn:
                        nummod = k
                ds_Ts = xr.open_dataset(fn)
                ds_Ts.close()
                ds_Ts.coords['lon'] = np.mod(ds_Ts.coords['lon'] + 180,360) - 180
                # select only region of interest (large)
                ds_Ts = ds_Ts[i].sel(lat=slice(ldom[0,0],ldom[0,1]),lon=slice(ldom[1,0],ldom[1,1]))
                # mask qin region
                ds_Ts = ds_Ts*qin

                # average by decade
                for j in range(len(yrsav)):
                    # decadal average
                    tmp2 = ds_Ts.sel(time=slice(str(yrsav[j])+'-01-01',str(yrsav[j]+9)+'-12-31'))
                    tmp2 = tmp2.mean(dim='time', skipna=True)
                    # apply optimal range
                    tmp2 = tmp2.where((tmp2>=th1)&(tmp2<=th2), np.nan)

                    tmp1[:,:,j,nummod] = tmp2.values
                # recreate into a data set
            tmp_ds = xr.DataArray(data=tmp1,dims=['lat','lon','decade','model'],
                                    coords=dict(lon=('lon',ds_Ts.lon),lat=('lat',ds_Ts.lat),
                                                decade=yrsav, model=models))
            # create 1 dataset with both variables
            varn = i2+'_'+i
            if (i=='thetao') & (i2=='surface'):
                ds_bothvar = xr.Dataset({varn:tmp_ds})
            else:
                ds_bothvar[varn] = tmp_ds
                
            ds_bothvar.to_netcdf(ldp+'QIN/downscaled_data/'+fish+'_decadal_2variables2depths.nc')
                                  
    return ds_bothvar
```

```python
ds = suitable_habitat_qin_decadal_map('Sablefish')

# quantify the suitablle habitat time series
# decades to average
yrsav = [*range(1990,2100,10)]
var = ['thetao','o2']   
depth = ['surface','bottom']
# for each depth
for i2 in depth:
    # for each variable 
    for i in var:
        varn = i2+'_'+i
        # for each model
        tmp1 = np.full((len(yrsav),len(models)),np.nan)
        for modi,modnam in enumerate(models): 
            tmp = ds[varn].sel(model=modnam)
            # for each decade
            for j in range(len(yrsav)):
                if np.nansum(tmp[:,:,:])>0:
                    tmp1[j,modi] = np.round(tmp[:,:,j].count(dim=['lat','lon']).values*100/npx,1) 
        plt.figure(figsize=(8,4))
        plt.plot(yrsav,tmp1,'.-', zorder=2, c='grey', alpha=0.5)
        plt.plot(yrsav,np.nanmean(tmp1,axis=1),'d-', zorder=3, c='tab:red', alpha=0.9)
        plt.errorbar(yrsav,np.nanmean(tmp1,axis=1),yerr=np.nanstd(tmp1,axis=1),c='tab:red', alpha=0.9, zorder=3)
        plt.grid(True, alpha=0.3, zorder=1)
        plt.ylabel('% of suitable habitat')
        plt.title(fishes[0]+', '+i2+' '+i)
        #plt.savefig('../figures/suitable_habitat/'+fish+'+_ensemblesuitablehabitatdecadal_'+var+'_'+depth+'_timeseries.png')
        plt.show()
```
