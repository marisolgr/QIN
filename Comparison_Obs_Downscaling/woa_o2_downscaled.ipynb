```python
import xarray as xr
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import glob
import cartopy.crs as ccrs
import cartopy
import scipy.stats as stat
from cartopy.mpl.ticker import LongitudeFormatter, LatitudeFormatter
import warnings
warnings.simplefilter('ignore') #filter some warning messages
```

```python
# definitions
ld = '/Users/sol/Downloads/ocldb1612540805.5656.CTD/'
ldom = np.array([[43, 48.5],[-129, -123]]) 
```

```python
### Reading CTD data
lats = list()
lons = list()
fechas = list()
depth = list()
oxy = list()

fns = glob.glob(ld+'/*.nc')
for i in fns:
    ds = xr.open_dataset(i)
    ds.close()
    try: 
        o = ds.Oxygen[-4:-1].mean().values
        d = ds.z[-4:-1].mean().values
        la = ds.lat.values
        lo = ds.lon.values
        if (la>=ldom[0,0]) & (la<=ldom[0,1]) & (lo>=ldom[1,0]) & (lo<=ldom[1,1]):
            if (o>0) & (d>=10) & (d<=4000):
                oxy.append(o*0.022391)
                lats.append(la)
                lons.append(lo)
                fechas.append(ds.time.values)
                depth.append(d)
    except:
        print('no oxygen\n')

## arrange in a dataset
var = [lats,lons,depth,fechas,oxy]
varnm = ['lat','lon','depth','time','oxygen']

for ind,i in enumerate(var):
    nda = xr.DataArray(data=i, dims=['pts'],coords={'pts':range(len(lats))})
    if ind==0:
        nds = xr.Dataset(data_vars={varnm[ind]:nda})
    else:
        nds = nds.assign(variables={varnm[ind]:nda})
```

```python
# for each point in world atlas, find bottom depth on Glorys model o2 bottom data
glo = xr.open_dataset('/Users/sol/Documents/LargeData/QIN/Glorys_Model/GLORYS_timeseries_extrapolated_bottom_o2_216x300.nc')
glo.close()
bdep = list()
for i in range(len(lats)):
    tmp = glo.sel(lat=slice(nds.lat[i]-0.05,nds.lat[i]+0.05),
               lon=slice(360+nds.lon[i]-0.05,360+nds.lon[i]+0.05),
               time=slice(nds.time[i]-np.timedelta64(15,'D'),nds.time[i]+np.timedelta64(15,'D'))).mean()
    bdep.append(tmp.depth.values)
# add to dataset
nda = xr.DataArray(data=bdep, dims=['pts'],coords={'pts':range(len(lats))})
nds = nds.assign(variables={'bottom_depth':nda})

# get rid of points with too much difference between where the last measure was taken and the bottom
difdep = nds['depth'].values - nds['bottom_depth']
nda = xr.DataArray(data=difdep, dims=['pts'],coords={'pts':range(len(lats))})
nds = nds.assign(variables={'depth_diff':nda})
nds = nds.where(np.abs(nds.depth_diff)<=150, drop=True)

nds.to_netcdf('wo_oxy_ctd.nc')
nds
```

```python
ldp = '/Users/sol/Documents/LargeData/QIN/downscaled_data/ssp485/o2/bottom/'
fns = glob.glob(ldp+'/*.nc')

plt.figure(figsize=(12,7))
nsp=1
for fn in fns:
    ds_Os=xr.open_dataset(fn)
    ds_Os.close()
 
    # for each point in world atlas, find model data
    modo2 = list()
    for i in range(len(nds.lat.values)):
        tmp = ds_Os.sel(lat=slice(nds.lat[i]-0.05,nds.lat[i]+0.05),
                   lon=slice(360+nds.lon[i]-0.05,360+nds.lon[i]+0.05),
                   time=slice(nds.time[i]-np.timedelta64(15,'D'),nds.time[i]+np.timedelta64(15,'D'))).mean()
        modo2.append(tmp.o2.values)
        
    tmps = fn.split('/')
    tmps2 = tmps[-1].split('_')
    modname = tmps2[0]+'_'+tmps2[1]
    rho,ps=stat.spearmanr(nds['oxygen'].values,modo2, nan_policy='omit')
    lg = 'rho='+str(np.round(rho,2))+', p='+str(np.round(ps,3))
    
    plt.subplot(2,3,nsp)
    plt.plot([-1.1,7],[-1.1,7],'k-',alpha=0.6)
    plt.scatter(nds['oxygen'].values,modo2,20,nds['depth_diff'],label=lg, cmap='inferno')
    cb = plt.colorbar()
    cb.ax.set_ylabel('depth diff')
    plt.axhline(y=0,c='k', alpha=0.3)
    plt.axvline(x=0,c='k', alpha=0.3)
    plt.grid(True, alpha=0.3)
    plt.title(modname)
    plt.xlabel('World atlas O2')
    plt.ylabel('Downscaled model O2')
    plt.legend(loc=2, fontsize='small')
    nsp +=1
    
plt.tight_layout()
plt.show()
```
